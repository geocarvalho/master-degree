```
if (!requireNamespace("BiocManager", quietly = TRUE))
    install.packages("BiocManager")

BiocManager::install("minfi")
BiocManager::install("GEOquery")
BiocManager::install("IlluminaHumanMethylation450kmanifest")
BiocManager::install("IlluminaHumanMethylation450kanno.ilmn12.hg19")
```

```
library(minfi)
library(GEOquery)
```

```
getGEOSuppFiles("GSE72251")
untar("GSE72251/GSE72251_RAW.tar", exdir = "GSE72251/idat")
head(list.files("GSE72251/idat", pattern = "idat"))
```

```
idatFiles <- list.files("GSE72251/idat", pattern = "idat.gz$", full = TRUE)
sapply(idatFiles, gunzip, overwrite = TRUE)
```

```
rgSet <- read.metharray.exp("GSE72251/idat")
gc()
grSet <- preprocessQuantile(rgSet)
```

```
# Download beta-values
beta <- getBeta(grSet)
write.csv(file="GSE72251_bvalues.csv", x=beta)
```

```
# Download phenotype data
geoMat <- getGEO("GSE72251")
pD.all <- pData(geoMat[[1]])
write.csv(file="GSE72251_all_phenotype.csv", x=pD.all)
```
